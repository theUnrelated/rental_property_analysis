```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from tqdm import tqdm
df3 = pd.read_csv('resale_mar-2012-to-dec-2014.csv')
df3.head()
```

```python
for i in tqdm(range(len(df3))):
    df3['rooms'] = df3.iat[i,3][0]
df3
```

```python
X = df3[['rooms', 'floor_area_sqm', 'lease_commence_date']]
y = df3[['resale_price']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print(len(X_test))
```

```python
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_test)
```

```python
# Find the confusion matrix of the result
cm = confusion_matrix(y_pred, y_test)
print(cm)

# Find the accuracy score of the result
asr = accuracy_score(y_pred, y_test)
print(asr)
```

```python
print("Feature importances:\n{}".format(tree.feature_importances_))
```
